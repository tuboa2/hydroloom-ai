### Engineered Features

```python
import os
import sys
import importlib

# fixed: resolve the base project root (one level up from /scripts)
project_root = os.path.abspath(os.path.join(os.getcwd(), ".."))

# fixed: track down where the 'features' folder is actually hiding
possible_paths = [
    project_root,                            
    os.path.join(project_root, "src"),       
    os.path.join(project_root, "hydromind")  
]

selected_path = None
for path_option in possible_paths:
    if os.path.exists(os.path.join(path_option, "features")):
        selected_path = path_option
        break

# fixed: inject the correct path to the front of sys.path
if selected_path:
    for path_option in possible_paths:
        if path_option in sys.path:
            sys.path.remove(path_option)
    sys.path.insert(0, selected_path)
    print(f"Target layout found! Injected path: {selected_path}")
else:
    print("Could not locate a 'features' directory structurally. Defaulting to project root.")
    sys.path.insert(0, project_root)

# 4. Safe import and execution
import run_feature_engineering
importlib.reload(run_feature_engineering)

run_feature_engineering.run(hemisphere="north")
run_feature_engineering.run(hemisphere="south")
```
